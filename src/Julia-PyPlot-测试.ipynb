```julia
using PyPlot
```

```julia
# 垂直条形图
x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 8, 16]
xlabel("x")
ylabel("y")
title("bar plot")
bar(x, y, color="blue")
```

```julia
# 简单折线图
x = 1:50
xlabel("x")
y = rand(50)
ylabel("y")
title("base plot")
grid("on")
plot(x, y)
```

```julia
using PyPlot

# 手绘风格的图表
x = 1:10
xlabel("x")
y = ones(10)
for i = 1:1:10 
    y[i] = -i*i*i
end
ylabel("y")
title("XKCD plot")
# 更改图表效果
xkcd()
plot(x, y)
```

```julia
# 水平条形图
x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 8, 16]
xlabel("X")
ylabel("Y")
title("bar plot")
barh(x, y, color = "green")
```

```julia
# 饼图
labels = ["A";"B";"C";"D"]
colors = ["orange";"blue";"red";"green"]
sizes = [200;900;30;1400]
fig = figure("pyplot_piechart", figsize=(6, 6))
p = pie(sizes, labels = labels, shadow = true, 
	startangle = 90, colors=colors)
title("pie plot")
```

```julia
# 散点图
fig = figure("scatterplot", figsize = (6,6))
x = rand(50)
y = rand(50)
areas = 10000*rand(50);
scatter(x, y, s=areas, alpha=0.5)
grid("on")
xlabel("x")
ylabel("y")
title("Scatter plot")
```

```julia
# 直方图
x = rand(100)
y = rand(100)
xlabel("x")
ylabel("y")
title("hist2D plot")
hist2D(x, y, bins=10)
```

```julia
# 3D圆柱
a = LinRange(0.0, 10pi, 3000)
b = LinRange(0.0, 6pi, 3000)

len_a = length(a)
len_b = length(b)

x = ones(len_a, len_b)
y = ones(len_a, len_b)
z = ones(len_a, len_b)

for i=1:len_a
    for j=1:len_b
        x[i,j] = sin(a[i])
        y[i,j] = cos(a[i])
        z[i,j] = sin(b[j])
    end
end

colors = rand(len_a, len_b, 3)
fig = figure()
surf(x, y, z, facecolors=colors)
fig[:canvas][:draw]()
```
